## Data Transformation
Text data requires preparation before it can be used in predictive modeling. We need to perform tokenization and vectorization of the text data.

#### Tokenization:
Tokenization is the process of breaking up text into words, keywords, phrases or other elements called tokens.
Tokens can be individual words, phrases or even whole sentences. 

#### Vectorization
Vectorization refers to encoding words or tokens as integers or floating point values for use as an input to a machine learning algorithm.
